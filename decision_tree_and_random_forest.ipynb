```python
#defines fit and transform function in DataFrameImputer class
#such that missing value is replaced by mean if the column values are numbers, 
#and by most-frequent element if the column values are strings.

import pandas as pd
import numpy as np
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.
        Columns of dtype object are imputed with the most frequent value 
        in column.
        Columns of other types are imputed with mean of column.
        """
    def fit(self, X, y=None):
        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)
        return self
    def transform(self, X, y=None):
        return X.fillna(self.fill)
```

```python
import math
from __future__ import division

class Node:
    
    nodeCount=0
    path_list=[]
    def __init__(self, data, labels, g_child=None, l_child=None, 
                                  leaf=False, label=None, split_info=None, depth=0):
        self.data=data
        self.labels=labels
    
        self.leaf=leaf
        self.label=label
        self.depth=depth
        
        self.g_child=g_child
        self.l_child=l_child
        self.split_info=split_info
        #self.max_depth=max_depth
        Node.nodeCount +=1

    #function that computes entropy
    def entropy(self, arr):  #array of 1s and 0s
        if(len(arr)==0):
            return 0
        ones = arr.count(1)/len(arr)
        zeros = arr.count(0)/len(arr)
        if ones==0 or zeros==0:
            return 0
        return -(ones*math.log(ones, 2) + zeros*math.log(zeros,2))
    
    #function that computes information gain
    def info_gain(self, p_arr, r_arr, l_arr):
        H_before = self.entropy(p_arr)
        H_after = (self.entropy(r_arr)*len(r_arr) + self.entropy(l_arr)*len(l_arr))/(len(r_arr)+len(l_arr))
        return H_before-H_after    
    
    #function that finds the best split. returns best splite feature, value and corresponding info gain
    def segmenter(self):
        lab_list=list(self.labels)
        H_before=self.entropy(lab_list)
        col_num=self.data.shape[1]
        max_gain_value=0
        max_split=0
        max_feature_col=0
        for i in range(col_num):
            uniq_list=np.unique(self.data[:,i])
            uniq_list=uniq_list[:-1]
            for j in uniq_list:
                ind_g = np.where(self.data[:,i]>j)
                ind_l = np.where(self.data[:,i]<=j)
                lab_g=list(np.take(self.labels, list(ind_g[0])))
                lab_l=list(np.take(self.labels, list(ind_l[0])))
                gain=self.info_gain(list(self.labels), lab_g, lab_l)
                if gain>max_gain_value:
                    max_gain_value=gain
                    max_split=j
                    max_feature_col=i
        self.split_info=[max_feature_col, max_split]
        return [max_feature_col, max_split, max_gain_value]
    
    def data_divider(self):
        col=self.segmenter()[0]
        ind_g = np.where(self.data[:,col]>self.segmenter()[1])
        ind_l = np.where(self.data[:,col]<=self.segmenter()[1])
        lab_g = np.take(self.labels, list(ind_g[0]))
        lab_l = np.take(self.labels, list(ind_l[0]))
        data_g = self.data[[list(ind_g[0])]]
        data_l = self.data[[list(ind_l[0])]]
        return data_g, lab_g, data_l, lab_l
    
    #a node is a leaf if it reaches a certain depth and/or
    #at least 5% of its elements are from same class
    def isLeaf(self):
        lab_list=list(self.labels)
        if(len(lab_list)==0):
            return False
        if self.depth==9:     #self.max_depth
            return True
        rat=lab_list.count(lab_list[0])/len(lab_list)
        if rat>=0.95 or rat<=0.05:
            return True
        else:
            return False
    #predictade label of a node
    def pred_label(self):
        if(self.isLeaf==False):
            return "None"
        lab_list=list(self.labels)
        
        if(len(lab_list)==0):
            return "None"
        n=lab_list[0]
        rat=(lab_list.count(n)/len(lab_list))
        if rat>=0.5:
            return n
        else:
            return (1-n)
    #recursively grows a tree
    def grow_tree(self):
        if(self.isLeaf()):
            return self.pred_label()

        self.depth +=1
        self.g_child=Node(data=self.data_divider()[0], labels=self.data_divider()[1], depth=self.depth)
        self.g_child.grow_tree()
        self.l_child=Node(data=self.data_divider()[2], labels=self.data_divider()[3], depth=self.depth)
        self.l_child.grow_tree()
    #given a test/valid point, it recursively finds the predicted class
    def predict(self, data):
        if self.isLeaf():
            return self.pred_label(), Node.path_list
        else:
            Node.path_list.append(self.split_info)
            if(data[self.split_info[0]]>self.split_info[1]):
                return self.g_child.predict(data)
            elif(data[self.split_info[0]]<=self.split_info[1]):
                return self.l_child.predict(data)            
```
